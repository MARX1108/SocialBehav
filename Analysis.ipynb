```python
#demo_yrs_1, demo_prnt_marital_v2, demo_comb_income_v2, demo_yrs_2

def preprocess():
    df = pd.read_csv("data/alldata.csv")
    #remove 777
    df = df[~(df == 777).any(axis=1)]
    #remove 999
    df = df[~(df == 999).any(axis=1)]

    df = df[~df.isnull().values.any(axis=1)]
    print(df.shape)
    df.to_csv('cleandata.csv')
    return df

preprocess()
```

```python
df = preprocess()

df = df.drop(['prosocial_child', 'prosocial_parent', 'interview_date', 'interview_age', 'subjectkey', 'sex'], axis=1)

from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats

X = np.array(df[df.columns[1:]])
y = np.array(df['aggressive_sumscore'])
reg = LinearRegression().fit(X, y)
print(df[df.columns[1:]].columns)
print('R2: ', reg.score(X, y))
print('coef: ', reg.coef_)
#print(reg.summary())

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
